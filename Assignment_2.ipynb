```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Linear Reagression**

```python
rng = np.random.RandomState(30)
x = rng.rand(50)
y = 2*x + rng.randn(50)
plt.plot(x,y,"go")
```

```python
def mean(y):
    return np.sum(y)/len(y)
```

```python
def loss1(y_calc,y_actual):
    return (np.mean((np.abs(y_calc - y_actual))**3))
```

```python
def hypothesis(theta_0,theta_1, x):
    return (theta_0 + theta_1*x)
```

```python
theta_0 = np.random.randn(1)
theta_1 = np.random.randn(1)
L =[]
learningRate = 0.01
iterations = 40000
for i in range(0,iterations):
    dtheta_0 = np.sum(np.multiply(3*(theta_0 + theta_1*x - y)**2,(np.sign(theta_0 + theta_1*x - y))))
    dtheta_1 = np.sum(np.multiply(3*np.multiply((theta_0 + theta_1*x - y)**2,x),(np.sign(theta_0 + theta_1*x - y))))
    theta_0 = theta_0 - learningRate/50 * dtheta_0
    theta_1 = theta_1 - learningRate/50 * dtheta_1
    L.append(loss1(hypothesis(theta_0,theta_1,x),y))
    if i%3000 == 0:
        plt.plot(x,y,"go")
        plt.plot(x,hypothesis(theta_0,theta_1,x), "ro")
        print(theta_0, theta_1)
        plt.show()
```

```python
plt.plot(L)
```

```python
def loss2(y_calc, y_actual):
    return (np.mean(np.abs(y_calc - y_actual)))
```

```python
theta_0 = np.random.randn(1)
theta_1 = np.random.randn(1)
L =[]
learningRate = 0.015
iterations = 40000
for i in range(0,iterations):
    dtheta_0 = np.sum(np.sign(hypothesis(theta_0,theta_1,x) - y))
    dtheta_1 = np.sum(np.multiply(x,np.sign(hypothesis(theta_0,theta_1,x) - y)))
    theta_0 = theta_0 - learningRate/50 * dtheta_0
    theta_1 = theta_1 - learningRate/50 * dtheta_1
    L.append(loss2(hypothesis(theta_0,theta_1,x),y))
    if i%3000 == 0:
        plt.plot(x,y,"go")
        plt.plot(x,hypothesis(theta_0,theta_1,x), "ro")
        print(theta_0, theta_1)
        plt.show()
```

```python
plt.plot(L)
```

**Polynomial Regression**

```python
rng = np.random.RandomState(20)
x = rng.rand(50)
y = 2*x + 5*x**2 +2*rng.randn(50)
plt.plot(x,y,"go")
```

```python
def hypothesis(theta_0, theta_1, theta_2, x):
    return (theta_0) + (theta_1)*x + (theta_2)*np.square(x)
```

```python
def loss3(y_calc, y_actual):
    return np.mean((y_calc-y_actual)**4)
```

```python
theta_0 = np.random.randn(1)
theta_1 = np.random.randn(1)
theta_2 = 5*np.random.randn(1)
sum = np.sum(y)
s= y/sum
z = hypothesis(theta_0,theta_1,theta_2,x)/sum
learningRate = 0.1
iterations = 80000
for i in range(0,iterations):
    dtheta_0 = np.sum(4*np.power(z - s,3))*100/50
    dtheta_1 = np.sum(4*np.multiply(np.power(z - s,3), x))*100/50
    dtheta_2 = np.sum(4*np.multiply(np.power(z - s,3), x**2))*100/50
    theta_0 -= learningRate*dtheta_0
    theta_1 -= learningRate*dtheta_1
    theta_2 -= learningRate*dtheta_2
    z = hypothesis(theta_0,theta_1,theta_2,x)/sum
    if i%3000 == 0:
        plt.plot(x,y,"go")
        plt.plot(x,hypothesis(theta_0,theta_1,theta_2,x), "ro")
        print(theta_0, theta_1,theta_2)
        plt.show()
```

```python
def loss4(y_calc, y_actual):
    return np.abs((y_calc-y_actual)**7)
```

```python
theta_0 = (np.random.randn(1))
theta_1 = (np.random.randn(1))
theta_2 = (np.random.randn(1))
sum = np.sum(y)
print(theta_0,theta_1,theta_2)
s = y/sum
z = hypothesis(theta_0,theta_1,theta_2,x)
z = z/sum
plt.show()
L = []
learningRate = 0.001
iterations = 80000
for i in range(0,iterations):
    dtheta_0 = np.sum(7*np.multiply(np.power(z - s,6),np.sign(z-s)))*(10**10)/50
    dtheta_1 = np.sum(7*np.multiply(np.multiply(np.power(z - s,6), x),np.sign(z-s)))*(10**10)/50
    dtheta_2 = np.sum(7*np.multiply(np.multiply(np.power(z - s,6), x**2),np.sign(z-s)))*(10**10)/50
    theta_0 -= learningRate*dtheta_0
    theta_1 -= learningRate*dtheta_1
    theta_2 -= learningRate*dtheta_2
    z = hypothesis(theta_0,theta_1,theta_2,x)
    z = z/sum
    L.append(np.mean(loss4(z,s)))
    if i%3000 == 0:
        print(theta_0,theta_1,theta_2)
        plt.plot(x,y,"go")
        plt.plot(x,z*sum, "ro")
        plt.show()
```
